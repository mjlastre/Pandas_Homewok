```python
import pandas as pd
import numpy as np 
```

```python
file_to_load = "../Resources/purchase_data.csv"
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
#purchase_data_df['SN'].nunique()
Total_players=purchase_data_df['SN'].nunique()
print("Total Number of Players is " + str (purchase_data_df['SN'].nunique())+".")
```

```python
Number_Unique= purchase_data_df["Item ID"].nunique()
Average_Price=purchase_data_df["Price"].mean()
Number_Purchases=purchase_data_df["Purchase ID"].count ()
Total_Revenue= purchase_data_df["Price"].sum ()
Summary=pd.DataFrame({"Number of Unique Items":[Number_Unique],"Average Price":[Average_Price],"Number of Purchase":[Number_Purchases],"Total Revenue":[Total_Revenue]})
Summary["Average Price"]=Summary["Average Price"].map("${:,.2f}".format)  
Summary["Total Revenue"]=Summary["Total Revenue"].map("${:,.2f}".format)  
Summary
#print ("Number of Unique Items is " + str(purchase_data_df["Item ID"].nunique())+ ".")
#print ("Average Price is $" + str(round(purchase_data_df["Price"].mean(),2)))
#print ("Number of Purchases is  "+  str(purchase_data_df["Purchase ID"].count ())+ "." )
#print ("Total Revenue is $ "+  str(purchase_data_df["Price"].sum ()))
```

```python
# Purchasing Analysis (Total)

Gender_df=purchase_data_df.groupby("Gender")["SN"]
#print(Gender_df.count().head(3))
Total_Count =Gender_df.nunique()
Percentage= Total_Count/Total_players*100
summary_table=pd.DataFrame({"Total Count":Total_Count,"Percentage of Players":Percentage})
summary_table
```

```python
# Purchasing Analysis (Gender)

puchased_count=purchase_data_df.groupby("Gender")
Purchase=puchased_count["Gender"].count() #count
Avg_Price=puchased_count["Price"].mean() # Average Purchase Price
Total_Purchase=puchased_count["Price"].sum()
Avg_Purchased=Total_Purchase/Total_Count #Avg Total Purchase per Person
Summary2=pd.DataFrame({"Purchase Count":Purchase,"Average Purchase Price":Avg_Price,"Total Purchase Value":Total_Purchase,"Avg Total Purchase per Person":Avg_Purchased})
Summary2["Average Purchase Price"]=Summary2["Average Purchase Price"].map("${:,.2f}".format)  
Summary2["Total Purchase Value"]=Summary2["Total Purchase Value"].map("${:,.2f}".format) 
Summary2["Avg Total Purchase per Person"]=Summary2["Avg Total Purchase per Person"].map("${:,.2f}".format) 
Summary2
```

```python
# Age Demographics

ageDemo_df=purchase_data_df
age_group=["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
bins=[0,9,14,19,24,29,34,39,50]
ageDemo_df["Age Group"]=pd.cut(ageDemo_df["Age"],bins,labels=age_group)
ageDemo_df=ageDemo_df.groupby("Age Group")
Total_Count1=ageDemo_df["SN"].nunique()
PerofTotal=round (Total_Count1/Total_players*100,2)
Summary3=pd.DataFrame({"Total Count":Total_Count1,"Percentage of Players":PerofTotal})
Summary3
```

```python
#Purchasing Analysis (Age)
ageDemo_df
Purchase=ageDemo_df["Price"].count() 
Avg_Price=ageDemo_df["Price"].mean() # Average Purchase Price
Total_Purchase=ageDemo_df["Price"].sum()
Avg_Purchased=Total_Purchase/Total_Count1 #Avg Total Purchase per Person
Summary4=pd.DataFrame({"Purchase Count":Purchase,"Average Purchase Price":Avg_Price,"Total Purchase Value":Total_Purchase,"Avg Total Purchase per Person":Avg_Purchased})
Summary4["Average Purchase Price"]=Summary4["Average Purchase Price"].map("${:,.2f}".format)  
Summary4["Total Purchase Value"]=Summary4["Total Purchase Value"].map("${:,.2f}".format) 
Summary4["Avg Total Purchase per Person"]=Summary4["Avg Total Purchase per Person"].map("${:,.2f}".format) 
Summary4
```

```python
# Top Spenders
TopSpenders_df=purchase_data_df.groupby(["SN"])
top=TopSpenders_df["Purchase ID"].count() # Purchase Count
Average= TopSpenders_df["Price"].sum()/top #Average Purchase Price
Total_=TopSpenders_df["Price"].sum()
Summary5=pd.DataFrame({"Purchase Count":top,"Average_Purchase_Price":Average,"Total_Purchase_Value":Total_})
Summary5["Average_Purchase_Price"]=Summary5["Average_Purchase_Price"].map("${:,.2f}".format)
Summary5=Summary5.sort_values("Total_Purchase_Value",ascending=False)
Summary5.iloc[:5]

```

```python
#Most Popular Items
MostPopular_df=purchase_data_df.groupby(["Item ID", "Item Name"])
Count=MostPopular_df["SN"].nunique() #Purchase Count
PriceI=MostPopular_df["Price"].mean() #Item Price
TotalP=MostPopular_df["Price"].sum() #Total Purchase Value
Summary6=pd.DataFrame({"Purchase Count":Count,"Item Price":PriceI,"Total Purchase Value":TotalP})
Summary6=Summary6.sort_values("Purchase Count",ascending=False)
Summary6
```

```python
#Most Profitable Items
Summary7=pd.DataFrame({"Purchase Count":Count,"Item Price":PriceI,"Total Purchase Value":TotalP})
Summary7=Summary7.sort_values("Total Purchase Value",ascending=False)
Summary7
```
